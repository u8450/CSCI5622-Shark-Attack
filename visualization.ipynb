```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("small_shark_data_clean.csv")
df.head(5)
```

```python
# box plot
plt.boxplot(df['Age'])
plt.title('Age of Shark Attack Victims')
plt.ylabel('Age (years)')
plt.show()
```

```python
df['Age'].describe()
```

```python
male = df[df["Sex"] == "M"]
plt.boxplot(male['Age'])
plt.title('Age of Male Shark Attack Victims')
plt.ylabel('Age (years)')
plt.show()
```

```python
male['Age'].describe()
```

```python
female = df[df["Sex"] == "F"]
plt.boxplot(female['Age'])
plt.title('Age of Female Shark Attack Victims')
plt.ylabel('Age (years)')
plt.show()
```

```python
female['Age'].describe()
```

```python
# pie chart
mylabels = ["M", "F"]
plt.pie(df['Sex'].value_counts(), labels=mylabels, explode = [0.2, 0])
plt.title('Gender of Shark Attack Victims')
plt.show()
```

```python
# pie chart
plt.plot(df['Year'].value_counts().sort_index())
plt.title('Amount of Shark Attack Victims by Year')
plt.xlabel('Year')
plt.ylabel('Count of shark attack')
plt.show()
```

```python
# Most attack prone activity
counts = df['Activity'].value_counts()
counts[0:5].plot(kind="bar")
plt.title('Top 5 activities with Shark Attack Victims')
plt.xlabel('Activities')
plt.ylabel('Count of shark attack')
```

```python
# Least attack prone activity
counts = df['Activity'].value_counts()
counts[-5:].plot(kind="bar")
plt.title('Bottom 5 activities with Shark Attack Victims')
plt.xlabel('Activities')
plt.ylabel('Count of shark attack')
```

```python
# States with most attacks
state = df['Area'].value_counts()
state[0:5].plot(kind="bar")
plt.title('Top 5 States with Most Shark Attack')
plt.xlabel('State')
plt.ylabel('Count of shark attack')
```

```python
# States with most fatal attacks
fatal = df[df['Fatal (Y/N)'] == "Y"]
state = fatal['Area'].value_counts()
state[0:5].plot(kind="bar")
plt.title('Top 5 States with Most Fatal Shark Attack')
plt.xlabel('State')
plt.ylabel('Count of Fatal Shark Attack')
```

```python
# Pie chart of fatal vs non fatal attacks

mylabels = ["Nonfatal", "Fatal"]
plt.pie(df['Fatal (Y/N)'].value_counts(), labels=mylabels, explode = [0.2, 0])
plt.title('Fatality of Shark Attacks')
plt.show()
```
